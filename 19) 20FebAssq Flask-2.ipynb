##### Q1. Explain GET and POST methods.

In Flask, **GET** method retrieves data from the server through URL parameters, while **POST** method sends data to the server through a request body.

##### Q2. Why is request used in Flask?

In Flask, the `request` object is used to access incoming request data such as form data, JSON, files, and more, allowing the Flask application to interact with the data sent by a client's request.

##### Q3. Why is redirect() used in Flask?

The `redirect()` function in Flask is used to redirect the user's browser to a different URL, typically after processing a form submission or upon certain conditions in a web application.

##### Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are files containing HTML and placeholders for dynamic content that are used to generate web pages dynamically.    
The `render_template()` function in Flask is used to render these templates, allowing dynamic content to be inserted into the HTML before sending it to the client's browser.

##### Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask,request,jsonify

app = Flask(__name__)

@app.route('/postman_data', methods=['POST'])
def math_calculation():
    if(request.method=='POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])

        if (ops=='add'):
            result = f"{num1} + {num2} = {num1+num2}"
        
        if (ops=='subtract'):
            result = f"{num1} - {num2} = {num1+num2}"

        if (ops=='multiply'):
            result = f"{num1} X {num2} = {num1*num2}"

        if (ops=='divide'):
            result = f"{num1} / {num2} = {num1/num2}"

        if (ops=='power'):
            result = f"{num1} raised to Power {num2} = {num1**num2}"

        return jsonify(result)

if __name__==('__main__'):
    app.run(host='0.0.0.0')
```
